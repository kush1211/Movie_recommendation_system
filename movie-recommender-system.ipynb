```python
import numpy as np
import pandas as pd
```

```python
movies = pd.read_csv('/kaggle/input/movies/tmdb_5000_movies.csv')
movies
```

```python
credits = pd.read_csv('/kaggle/input/movies/tmdb_5000_credits.csv')
credits
```

```python
movies = movies.merge(credits,on = 'title')
movies
```

```python
movies = movies[['movie_id','title','overview','genres','keywords','cast','crew']]
movies
```

```python
movies['crew'].iloc[2]
```

```python
import ast

def convert(obj):
    L = []
    for i in ast.literal_eval(obj):
        L.append(i['name'])
    return L

def convert_cast(obj):
    L = []
    count = 0
    for i in ast.literal_eval(obj):
        if count!=10:
            L.append(i['name'])
            count+=1
    return L

def convert_crew(obj):
    L = []
    for i in ast.literal_eval(obj):
        if i['job']=='Director':
            L.append(i['name'])
    return L
```

```python
movies['genres'] = movies['genres'].apply(convert)
movies['keywords'] = movies['keywords'].apply(convert)
movies
```

```python
movies['cast'] = movies['cast'].apply(convert_cast)
movies['crew'] = movies['crew'].apply(convert_crew)
movies
```

```python
movies.isna().sum()
```

```python
movies.dropna(inplace=True)
```

```python
movies
```

```python
def collapes(list):
    L =[]
    for i in list:
        L.append(i.replace(' ',''))
    return L
    
```

```python
movies['genres'] = movies['genres'].apply(collapes)
movies['keywords'] = movies['keywords'].apply(collapes)
movies['cast'] = movies['cast'].apply(collapes)
movies['crew'] = movies['crew'].apply(collapes)
movies
```

```python
movies['tags'] = movies['overview'].apply(lambda x:x.split(' ')) + movies['genres'] + movies['keywords'] + movies['cast'] + movies['crew']
movies
```

```python
movies.drop(['overview','genres','keywords','cast','crew'],axis=1,inplace = True)
movies
```

```python
movies['tags'] = movies['tags'].apply(lambda x:' '.join(x))
movies
```

```python
movies['tags'] = movies['tags'].str.lower()
movies
```

```python
import spacy
nlp = spacy.load('en_core_web_sm')

def lemma(text):
    ans=[]
    tokens = nlp(text)
    for token in tokens:
        ans.append(token.lemma_)
    return ' '.join(ans)
```

```python
movies['tags'] = movies['tags'].apply(lemma)
```

```python
movies
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer

tfidf_vectorizer = TfidfVectorizer(max_features=5000,stop_words='english')

tfidf_matrix = tfidf_vectorizer.fit_transform(movies['tags'])

feature_names = tfidf_vectorizer.get_feature_names_out()

feature_names
```

```python
from sklearn.metrics.pairwise import cosine_similarity
similarity = cosine_similarity(tfidf_matrix)
```

```python
similarity
```

```python
similarity.shape
```

```python
def recommend(title):
    index = movies[movies['title'] == title].index[0]
    distances = sorted(list(enumerate(similarity[index])),reverse = True,key = lambda x:x[1])
    for i in distances[1:6]:
        print(movies.iloc[i[0]].title)
    
```

```python
recommend('Avatar')
```

```python
recommend('Batman Begins')
```
